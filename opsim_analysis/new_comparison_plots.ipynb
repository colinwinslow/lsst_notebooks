```python
from __future__ import print_function
import lsst.sims.maf.plots as plots
%matplotlib inline
```

```python
from lsst.sims.maf.runComparison import RunComparison
```

```python
runnames = ['astro-lsst-01_2013',
            'astro-lsst-01_2020',
            'minion_1016_newsky']
```

```python
comp = RunComparison(baseDir='.', runlist=runnames, verbose=True)
```

```python
bundleDict, mname = comp.readMetricData('Median airmass',
                                        metricMetadata='r band, all props',
                                        slicerName="OpsimFieldSlicer")
```

```python
comp.plotMetricData(bundleDict, plotFunc=plots.BaseSkyMap(), 
                    userPlotDict={'figsize':(15,15)})
```

```python
comp.plotMetricData(bundleDict, plotFunc=plots.OpsimHistogram(), 
                    userPlotDict={'figsize':(8,8),'binsize':0.05, 'legendloc':1})
```
